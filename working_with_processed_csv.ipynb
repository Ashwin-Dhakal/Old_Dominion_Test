```python
from wordcloud import WordCloud
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("processed_data.csv")
```

```python
df.head()
```

```python
#Checking out the most common words in word cloud graph from the title.

cloud1 = WordCloud(width=1280, height=980).generate(" ".join(df['TITLE'].astype(str)))

plt.figure(figsize=(15, 10))
plt.imshow(cloud1)
plt.axis('off')
```

```python
#Checking out the most common words in word cloud graph from the abstract.

cloud2 = WordCloud(width=1280, height=980).generate(" ".join(df['PROCESSED_ABSTRACT'].astype(str)))

plt.figure(figsize=(15, 10))
plt.imshow(cloud2)
plt.axis('off')

```
